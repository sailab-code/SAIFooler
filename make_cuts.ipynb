```python
from PIL import Image
from IPython.display import display
```

```python
crop_boxes = [
    (680, 454),
    (906, 1132)
]
size = 223
```

```python
pref = "C:/Users/enric/OneDrive/Universit√†/Dottorato/SAILenv/saifooler/eps0.1/remote/remote"
img_pre = Image.open(f"{pref}_pre_sailenv.png")
img_post = Image.open(f"{pref}_post_sailenv.png")
collage = Image.new(mode="RGB", size=(size*2, size*2))
```

```python

for box_idx, crop_box in enumerate(crop_boxes):
    box = (crop_box[0], crop_box[1], crop_box[0] + size, crop_box[1] + size)
    crop_pre = img_pre.crop(box)
    crop_post = img_post.crop(box)
    display(crop_pre)

    row = box_idx * size

    # paste pre
    box_pre = (0, row, size, row + size)
    collage.paste(crop_pre, box_pre)

    # paste post
    box_pre = (size, row, 2 * size, row + size)
    collage.paste(crop_post, box_pre)
    
collage
```

```python
collage.save(f"{pref}_out_sailenv.png")
```
